## **Web Scraping**

* Utilizing BeautifulSoup and Selenium, the script gathers employment details from the Naukri.com website.
* It retrieves information such as job titles, company names, experience prerequisites, locations, and salary details.

```python
# Importing necessary libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
# Open a Chrome browser window using Selenium WebDriver
# Accessing the specified URL using the WebDriver
driver = webdriver.Chrome()
driver.get('https://www.naukri.com/data-science-jobs?k=data%20science&nignbevent_src=jobsearchDeskGNB')
```

```python
# Explicitly wait for the title elements to be present on the page
wait = WebDriverWait(driver, 10)
titles = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'title')))
```

```python
# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
soup
```

```python
# Extracting all anchor tags with the class 'title'
soup1 = soup.find_all('a', class_='title')
```

```python
# Extract text content from each element in the list of anchor tags
list1 = [i.text.strip() for i in soup1]
```

```python
list1
```

```python
# Extracting all anchor tags with the class 'comp-name mw-25' from the BeautifulSoup object
soup2 = soup.find_all('a',class_ = 'comp-name mw-25')
```

```python
# Extracting text content from each element in the list of anchor tags
list2 = [i.text.strip() for i in soup2]
```

```python
list2
```

```python
# Extracting all span tags with the class 'expwdth' from the BeautifulSoup object
soup3 = soup.find_all('span', class_ = 'expwdth')
```

```python
# Extracting text content from each element in the list of span tags
list3 = [i.text.strip() for i in soup3]
```

```python
list3
```

```python
# Extracting text content from each element in the list of span tags with class 'locWdth'
soup4 = soup.find_all('span',class_ = 'locWdth')
```

```python
list4 = [i.text.strip() for i in soup4]
```

```python
list4
```

```python
# Extracting text content from each element in the list of span tags with class 'ni-job-tuple-icon ni-job-tuple-icon-srp-rupee sal'
soup5 = soup.find_all('span',class_ = 'ni-job-tuple-icon ni-job-tuple-icon-srp-rupee sal')
```

```python
list5 = [i.text.strip() for i in soup5]
```

```python
list5
```

```python
# Extracting text content from each element in the list of span tags with class 'ni-job-tuple-icon ni-job-tuple-icon-srp-rupee sal'
soup6 = soup.find_all('li',class_ = 'dot-gt tag-li')
```

```python
list6 = [i.text.strip() for i in soup6]
```

## **Creating Dataframe**


The **'create_dataframe'** function takes lists of job-related information as input and returns a DataFrame.

```python
def create_dataframe(titles, companies, experiences, locations, salaries, skills):
    # Determine the minimum length among all input lists
    min_length = min(len(titles), len(companies), len(experiences), len(locations), len(salaries), len(skills))

    # Use zip to iterate over lists simultaneously and truncate or pad as necessary
    data = {
        'Job Title': titles[:min_length],
        'Company Name': companies[:min_length],
        'Experience': experiences[:min_length],
        'Location': locations[:min_length],
        'Salary': salaries[:min_length],
        'Skill': skills[:min_length]
    }

    # Create a DataFrame from the processed data
    df = pd.DataFrame(data)
    return df

# Create a DataFrame using the provided lists
df = create_dataframe(list1, list2, list3, list4, list5, list6)
```

```python
# Print the dataframe
df
```

## **Visualization 1:** Number of Jobs per Company (Bar Chart)

The code visualizes the distribution of jobs across various companies by generating a bar chart. In this chart, each bar represents a company, and the height of the bar corresponds to the number of jobs available at that specific company. This visualization offers a rapid overview of the companies with the highest number of job listings.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the number of jobs per company
company_counts = df['Company Name'].value_counts()

# Plot a bar chart using seaborn
plt.figure(figsize=(10, 6))  # Adjust the figure size
sns.barplot(x=company_counts.index, y=company_counts, palette='viridis') 

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Set plot title and labels
plt.title('Number of Jobs per Company')
plt.xlabel('Company Name')
plt.ylabel('Number of Jobs')

# Display the plot
plt.show()
```

## **Visualization 2:** Distribution of Experience (Histogram)


The code generates a histogram to illustrate the distribution of experience levels among data science job listings. The green bars represent the frequency of jobs within various experience ranges, and the KDE plot provides a smooth estimate of the underlying probability density.

```python
# Set the figure size
plt.figure(figsize=(10, 8))

# Create a histogram using seaborn
sns.histplot(df['Experience'], bins=20, color='green', kde=True)

# Set the title and labels
plt.title('Distribution of Experience')
plt.xlabel('Experience')
plt.ylabel('Frequency')

# Add a legend for clarity
plt.legend(labels=['Experience Distribution'], loc='upper right', fontsize=10)

# Add grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

## **Visualization 3:** Distribution of Job Titles (Count Plot)

This line exhibits a count plot depicting the distribution of job titles in the data science job listings, with a specified figure size. The count plot visually represents the frequency of each job title, offering insight into the distribution of jobs in the field.

```python
# Set the figure size
plt.figure(figsize=(12, 10))

# Count plot with a customized color palette
sns.countplot(y='Job Title', data=df, order=df['Job Title'].value_counts().index, palette='viridis')

# Add title and labels with increased font size
plt.title('Distribution of Job Titles', fontsize=22)
plt.xlabel('Number of Jobs', fontsize=16)
plt.ylabel('Job Title', fontsize=16)

# Show the plot
plt.show()
```

## **Visualization 4:** Top 10 Skills in Demand (Horizontal Bar Chart)

This line showcases a horizontal bar chart illustrating the distribution of the top 10 in-demand skills across various job listings. The chart, with a specified figure size, visually represents the frequency of each skill, providing valuable insights into the prevailing skill requirements in the field.

```python
# Assuming you have a DataFrame named 'data' with columns: 'jobtitle' and 'skill'
top_skills = df['Skill'].value_counts().nlargest(10)

# Set the figure size
plt.figure(figsize=(12, 8))

# Count plot
sns.barplot(x=top_skills.values, y=top_skills.index)

# Add title and labels
plt.title('Top 10 Skills in Demand Across Job Titles', fontsize=22)
plt.xlabel('Number of Jobs', fontsize=16)
plt.ylabel('Skill', fontsize=16)

# Show the plot
plt.show()
```

## **Visualization 5:** Distribution of Joblistings Across Various Locations

This line presents a bar plot depicting the distribution of job listings across various locations. Each horizontal bar represents the frequency of job listings in a specific location, offering insights into the geographical distribution of job opportunities. The plot, created with a specified figure size, provides a visual representation of the job market landscape across different regions.

```python
# Set the figure size
plt.figure(figsize=(12, 10))

# Count plot with a customized color palette
sns.countplot(y='Location', data=df, order=df['Location'].value_counts().index, palette='viridis')

# Add title and labels with increased font size
plt.title('Distribution of Joblistings Across Locations', fontsize=22)
plt.xlabel('Frequency', fontsize=16)
plt.ylabel('Location', fontsize=16)

# Show the plot
plt.show()
```
